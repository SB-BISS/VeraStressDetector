```python
import requests
```

```python
url = 'http://127.0.0.1:8882/'
files = {'file': open('68118_78958_1541090078.mp3', 'rb')}

r = requests.post(url, files=files)
```

```python
import json
json_tuple = json.loads(r.text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

evalpd =  pd.DataFrame(json_tuple['emotions_series']).T
   
evalpd.plot()  

plt.title('Stress Score: ' + str(json_tuple['stress'][0]))
    
```
